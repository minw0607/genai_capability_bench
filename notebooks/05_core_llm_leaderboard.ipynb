# Core LLM Capability Leaderboard

This notebook aggregates the core capability results into a single model scorecard. It is the answer to: what can this model do, and how well can it do it?

```python
from pathlib import Path
import sys
import json
import pandas as pd

sys.path.insert(0, str(Path('..').resolve() / 'src'))

from genai_capability_bench.core.runner import run_from_config
from genai_capability_bench.reporting.tables import capability_leaderboard, summarize_results
from genai_capability_bench.reporting.plots import plot_capability_scores
from genai_capability_bench.reporting.executive_summary import generate_summary

CONFIG = Path('../configs/eval_core_demo.yaml')
```

## Run Unified Core Evaluation

```python
run_dir = run_from_config(CONFIG)
results = json.loads((run_dir / 'results.json').read_text())
df = pd.DataFrame(results)
df[['model_name', 'capability', 'category', 'task_id', 'score', 'passed']]
```

## Leaderboard

```python
leaderboard = capability_leaderboard(results)
display(leaderboard)
summary = summarize_results(results)
plot_capability_scores(summary, title='Core LLM Capability Scorecard');
```

## Executive Summary

```python
print(generate_summary(summary))
print(f'Artifacts saved to: {run_dir}')
```
