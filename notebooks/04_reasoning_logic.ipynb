# Reasoning and Logic

This notebook evaluates tasks where the model must derive an answer rather than simply retrieve a fact. The starter evaluator checks the final answer; future versions can add stepwise rubrics and contradiction checks.

```python
from pathlib import Path
import sys
import json
import pandas as pd

sys.path.insert(0, str(Path('..').resolve() / 'src'))

from genai_capability_bench.core.runner import run_from_config
from genai_capability_bench.reporting.tables import summarize_results
from genai_capability_bench.reporting.plots import plot_capability_scores

CONFIG = Path('../configs/eval_core_demo.yaml')
```

## Run Evaluation

```python
run_dir = run_from_config(CONFIG)
df = pd.DataFrame(json.loads((run_dir / 'results.json').read_text()))
reason_df = df[df['capability'] == 'reasoning_logic']
reason_df[['task_id', 'category', 'actual_output', 'expected_output', 'score', 'metrics', 'passed']]
```

## Reasoning Summary

```python
summary = summarize_results(reason_df.to_dict(orient='records'))
display(summary)
plot_capability_scores(summary, title='Reasoning and Logic');
```
