# Truthfulness

This notebook evaluates whether responses align more closely with correct references than with known incorrect or misconception-style references. It is the natural home for TruthfulQA-style workflows.

```python
from pathlib import Path
import sys
import json
import pandas as pd

sys.path.insert(0, str(Path('..').resolve() / 'src'))

from genai_capability_bench.core.runner import run_from_config
from genai_capability_bench.reporting.tables import summarize_results
from genai_capability_bench.reporting.plots import plot_capability_scores
from genai_capability_bench.reporting.executive_summary import generate_summary

CONFIG = Path('../configs/eval_core_demo.yaml')
```

## Run and Inspect Truthfulness Cases

```python
run_dir = run_from_config(CONFIG)
df = pd.DataFrame(json.loads((run_dir / 'results.json').read_text()))
truth_df = df[df['capability'] == 'truthfulness']
truth_df[['task_id', 'category', 'actual_output', 'expected_output', 'score', 'metrics', 'passed']]
```

## Summary

The current local metric is a correct-vs-incorrect similarity margin. Future production runs can add LLM-as-judge truthfulness rubrics using the shared judge helper.

```python
summary = summarize_results(truth_df.to_dict(orient='records'))
display(summary)
plot_capability_scores(summary, title='Truthfulness Summary');
print(generate_summary(summary))
```
